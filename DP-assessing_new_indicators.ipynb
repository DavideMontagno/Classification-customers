## Pil

Da inserire grafico che mostra il pil ossia la nazione

```python
df_pil = pd.read_csv("dataset/Country_PIL.csv", sep=";", decimal=",",index_col=0)
dict_pil={}
for index, row in df_pil.iterrows():
    dict_pil[row["PIL"]]=row["Country"]
```

```python
temp_pil=df.copy()
temp_pil["Coutry"]="N"
for index, row in temp_pil.iterrows():
    temp_pil.loc[index]=dict_pil[row["PIL"]]
```

```python
temp_pil.groupby("Coutry")["CustomerID"].count().plot.bar()
```
