## Correlation

```python
import seaborn as sns
df.corr()
fig, ax = plt.subplots(figsize=(15,10))         # Sample figsize in inches
heatmap = sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=True, linewidths=1.5, ax= ax)
heatmap.set_title('HeatMap correlazione nuovi indicatori', fontdict={'fontsize':12}, pad=10)
```

Le colonne da eliminare perchè dipendenti dal tempo sono:
* I
* Iu
* Ts

Le colonne da eliminare per l'alta correlazione sono:

* TsM o TsW
* NbxM o NbxW
* MaxD, MinD
* Etrim o Emonth o Eday_of_week o Ehour(lasciarne 1)

```python
import seaborn as sns
del df['I']
del df['Iu']
del df['Ts']
del df['TsW']
del df['NbxW']
del df['MaxD']
del df['MinD']
del df['Emonth']
del df['Eday_of_week']
del df['Ehour']
df.corr()
fig, ax = plt.subplots(figsize=(15,10))         # Sample figsize in inches
heatmap = sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=True, linewidths=1.5, ax= ax)
heatmap.set_title('HeatMap correlazione nuovi indicatori', fontdict={'fontsize':12}, pad=10)
```
