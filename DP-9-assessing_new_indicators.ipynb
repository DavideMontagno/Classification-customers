## Pil

Da inserire grafico che mostra il pil ossia la nazione

```python
df_pil = pd.read_csv("dataset/Country_PIL.csv", sep=";", decimal=",",index_col=0)
dict_pil={}
for index, row in df_pil.iterrows():
    dict_pil[row["PIL"]]=row["Country"]
```

```python
temp_pil=df.copy()
temp_pil["Coutry"]="N"
for index, row in temp_pil.iterrows():
    temp_pil.loc[index]=dict_pil[row["PIL"]]
```

```python
temp_pil.groupby("Coutry")["CustomerID"].count().plot.bar()
```

## Correlation

```python
import seaborn as sns
df.corr()
fig, ax = plt.subplots(figsize=(15,10))         # Sample figsize in inches
heatmap = sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=True, linewidths=1.5, ax= ax)
heatmap.set_title('HeatMap correlazione nuovi indicatori', fontdict={'fontsize':12}, pad=10)
```
